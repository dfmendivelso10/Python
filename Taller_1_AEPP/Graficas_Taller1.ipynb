## Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Modelo Teórico


```python

# Cargar el archivo de Excel
file_path = '/Users/daniel/Documents/Modelo_Teorico.xlsx'
df = pd.read_excel(file_path)
# Extraer las columnas relevantes para graficar
precio = df['precio']
oferta = df['Oferta']
demanda = df['Demanda']

# Graficar las curvas de oferta y demanda con el precio en el eje Y
plt.figure(figsize=(10, 6))
plt.plot(oferta, precio, label='Oferta', marker='o', linestyle='--', color='blue')
plt.plot(demanda, precio, label='Demanda', marker='o', linestyle='-', color='red')

# Agregar títulos y etiquetas
plt.title('Curvas de Oferta y Demanda')
plt.xlabel('Cantidad')
plt.ylabel('Precio')
plt.legend()

# Agregar una cuadrícula para mejor legibilidad
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo de Excel
file_path = '/Users/daniel/Documents/Modelo_Empirico.xlsx'
df2 = pd.read_excel(file_path)

# Extraer las columnas relevantes para graficar
precio = df2['precio']
oferta = df2['Oferta']
demanda = df2['Demanda']

# Crear la figura y el eje principal (oferta)
fig, ax1 = plt.subplots(figsize=(10, 6))

# Graficar la oferta en el eje Y principal
ax1.plot(oferta, precio, label='Oferta', marker='o', linestyle='--', color='blue')
ax1.set_xlabel('Cantidad de Oferta')
ax1.set_ylabel('Precio', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Crear un segundo eje Y (demanda) que comparte el mismo eje X
ax2 = ax1.twinx()  # twin Y-axis with the same X-axis
ax2.plot(demanda, precio, label='Demanda', marker='o', linestyle='-', color='red')
ax2.set_ylabel('Precio', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Agregar títulos y etiquetas
plt.title('Curvas de Oferta y Demanda')

# Agregar una cuadrícula para mejor legibilidad
ax1.grid(True)

# Mostrar la gráfica
plt.show()

```
